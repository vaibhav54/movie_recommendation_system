```python
import pandas as pd
import numpy as np
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv')
movies_df = pd.read_csv('tmdb_5000_movies.csv')
```

```python
movies_df.head()
```

```python
credits_column_renamed = credits.rename(index=str, columns={"movie_id":"id"})
movies_df_merge = movies_df.merge(credits_column_renamed, on='id')
movies_df_merge.head()
```

```python
movies_cleaned_df = movies_df_merge.drop(columns=['homepage','title_x','title_y','status','production_countries'])
movies_cleaned_df.head(1)['overview']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf = TfidfVectorizer(min_df=3, max_features=None, strip_accents='unicode', analyzer='word', token_pattern= r'\w{1,}',
                    ngram_range=(1,3), 
                    stop_words='english')
movies_cleaned_df['overview'] = movies_cleaned_df['overview'].fillna('')

```

```python
import sklearn
print(sklearn.__version__)
```

```python
tfv_matrix = tf.fit_transform(movies_cleaned_df['overview'])
```

```python
tfv_matrix
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel

#compares two different overview, and gives similarity between them inthe number between 0 and 1 
sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
sig[0]
```

```python
indices = pd.Series(movies_cleaned_df.index, index= movies_cleaned_df['original_title']).drop_duplicates()
```

```python
indices
```

```python
def give_recommendation(title, sig=sig):
    idx = indices[title]
    sig_scores = list(enumerate(sig[idx]))
    
    sig_scores = sorted(sig_scores, key=lambda x:x[1], reverse=True)
    
    sig_scores = sig_scores[1:11]
    
    movie_indices = [i[0] for i in sig_scores]
    
    return movies_cleaned_df['original_title'].iloc[movie_indices]
```
